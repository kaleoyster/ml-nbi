```python
import logistic_regression
import decision_tree
import svm
import gradient_boosting
import lg_boost
import xg_boost
import random_forest
import deep_learning

performance_log = logistic_regression.main()
performance_svm = svm.main()
performance_dt = decision_tree.main()
performance_gb = gradient_boosting.main()
performance_rf = random_forest.main()
performance_xgb = xg_boost.main()
performance_lg = lg_boost.main()
performance_dl = deep_learning.main()
```

## Accuracy values across machine learning model

```python
import plotly.graph_objects as go
import numpy as np

# Number of boxes
N = 8    

# Generate an array of rainbow colors by fixing the saturation and lightness of the HSL
c = ['hsl('+str(h)+',50%'+',50%)' for h in np.linspace(0, 360, 8)]

# Each box is represented by a dict that contains the data, the type, and the colour.
# Use list comprehension to describe N boxes, each with a different colour and with different randomly generated data:
fig = go.Figure(data=[
        go.Box(marker_color = c[0],
               name='Logistic Regression',
               y=np.array(performance_log['accuracy'])),
      go.Box(marker_color = c[1],
             name='Support Vector Machine',
             y=np.array(performance_svm['accuracy'])),
      go.Box(marker_color = c[2],
             name='Random Forest',
             y=np.array(performance_rf['accuracy'])),
      go.Box(marker_color = c[3],
             name='Decision Tree',
             y=np.array(performance_dt['accuracy'])),
      go.Box(marker_color = c[4],
             name='Gradient Boosting',
             y=np.array(performance_gb['accuracy'])),
      go.Box(marker_color = c[5],
             name='XGB',
             y=np.array(performance_xgb['accuracy'])),
      go.Box(marker_color = c[6],
             name='Light GB',
             y=np.array(performance_lg['accuracy'])),
    go.Box(marker_color = c[7],
             name='Deep learning',
             y=np.array(performance_dl['accuracy']))
])

# format the layout
fig.update_layout(
    title='Comparison of accuracy across machine learning models',
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=True),
    yaxis=dict(zeroline=True, gridcolor='white', title='Accuracy'),
    showlegend=False,
    paper_bgcolor='rgb(233,233,233)',
    plot_bgcolor='rgb(233,233,233)',
)

fig.show()
```

```python
import plotly.graph_objects as go
import plotly.express as px

fig = go.Figure()
fig.add_shape(
    type='line', line=dict(dash='dash'),
    x0=0,
    x1=1,
    y0=0,
    y1=1
)

models = ['logistic regression',
          'deep learning',
          'SVM',
          'Random Forest',
          'Decision Tree'
          'Gradient boosting',
          'XGB',
          'Light Boost'
         ]

auc_log = np.mean(performance_log['auc'])
auc_dl = np.mean(performance_dl['auc'])
auc_svm = np.mean(performance_svm['auc'])
auc_dt = np.mean(performance_dt['auc'])
auc_rf = np.mean(performance_rf['auc'])
auc_gb = np.mean(performance_gb['auc'])
auc_xgb = np.mean(performance_xgb['auc'])
auc_lg = np.mean(performance_lg['auc'])

fig.add_trace(go.Scatter(x=performance_log['fpr'][0],
                         y=performance_log['tpr'][0],
                         name=f'Logistic Regression, Mean AUC:{auc_log:.3f}',
                         mode='lines')
             )

fig.add_trace(go.Scatter(x=performance_dl['fpr'][0],
                         y=performance_dl['tpr'][0],
                         name=f'Deep Learning, Mean AUC:{auc_dl:.3f}',
                         mode='lines')
             )

fig.add_trace(go.Scatter(x=performance_svm['fpr'][0],
                         y=performance_svm['tpr'][0],
                         name=f'SVM, Mean AUC:{auc_svm:.3f}',
                         mode='lines')
             )

fig.add_trace(go.Scatter(x=performance_dt['fpr'][0],
                         y=performance_dt['tpr'][0],
                         name=f'Decision Tree, Mean AUC:{auc_dt:.3f}',
                         mode='lines')
             )

fig.add_trace(go.Scatter(x=performance_rf['fpr'][0],
                         y=performance_rf['tpr'][0],
                         name=f'Random forest, Mean AUC:{auc_rf:.3f}',
                         mode='lines')
             )



fig.add_trace(go.Scatter(x=performance_xgb['fpr'][0],
                         y=performance_xgb['tpr'][0],
                         name=f'XGBoost, Mean AUC:{auc_xgb:.3f}',
                         mode='lines')
             )


fig.add_trace(go.Scatter(x=performance_lg['fpr'][0],
                         y=performance_lg['tpr'][0],
                         name=f'Light Boost, Mean AUC:{auc_lg:.3f}',
                         mode='lines')
             )

fig.update_layout(
    xaxis_title='False Positive Rate',
    yaxis_title='True Positive Rate',
    yaxis=dict(scaleanchor="x", scaleratio=1),
    xaxis=dict(constrain='domain'),
    width=700, height=500
)
fig.show()
```
